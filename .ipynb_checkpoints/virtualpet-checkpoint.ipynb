The Goal: Create a Pet class with attributes like name, hunger, energy, and happiness. Create methods like feed(), play(), and sleep() that alter these attributes.
Concepts Tested: Classes, __init__, self, instance attributes, methods.


```python
class Pet:
    def __init__(self, name):
       self.name = name
       self.hunger = 50
       self.energy = 50
       self.happiness = 50
    def feed(self):
        print(f"{self.name} is enting")
        self.hunger -=20
        if self.hunger < 0:
            self.hunger = 0
        self.happiness += 0
    def play(self):
        print(f"{self.name} is playing")
        self.energy -=15
        self.hunger +=10
        self.happiness +=15
    def sleep(self):
        print(f"{self.name} is sleepping ")
        self.energy += 30
        if self.energy > 100:
            self.energy += 100
    def status(self):
        print(f"\n{self.name} is Status: ")
        print(f"Hunger: {self.hunger}")
        print(f"Enter energy: {self.energy}")
        print(f"Happiness: {self.happiness}")
        
```
