The Goal: A program that converts temperatures (Celsius to Fahrenheit), distances (Miles to Kilometers), and weights(Grams to Kilograms).
The Catch: Every single conversion must be its own independent function (e.g., def c_to_f(celsius):). There should be a main menu function that calls the others based on user choice.
Concepts Tested: Defining functions, passing arguments, return statements, variable scope.

```python
def c_to_f(celsius):
    return (celsius *9/5) + 32
def miles_to_km(miles):
    return miles * 1.60934
def grams_to_kg(grams):
    return grams / 1000
def main():
    while True:
        print(f"----Menu Converts----")
        print("1. Celsius to Fahrenheit")
        print("2. Miles to Kilometers")
        print("3. Grams to Kilograms")
        print("4. Exit")
        choice = input("Choice an Option(1-4): ")
        
        if choice == "1":
            c = float(input("Enter Celsius: "))
            result = c_to_f(c)
            print(f"Fahrenheit: {result:.2f}°F")
        elif choice == "2":
            m = float(input("Enter Miles: "))
            result = miles_to_km(m)
            print(f"Kilometers: {result:.2f}km")
        elif choice == "3":
            g = float(input("Enter Grams: "))
            result = grams_to_kg(g)
            print(f"Kilograms: {result:.2f}kg")
        elif choice == "4":
            print("Good by")
            break
        else :
            print("Invalid Choice.")
main()
```

The Goal: A function that takes an integer (length) and boolean flags (use_numbers, use_symbols) and returns a randomized, secure password.
Concepts Tested: Default arguments, keyword arguments, combining lists/strings, returning data.

```python
import random
import string

def generate_password(length, use_numbers=True, use_symbols=True):
    characters = string.ascii_letters  
    if use_numbers:
        characters += string.digits  
    if use_symbols:
        characters += string.punctuation 

    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return password

pwd = generate_password(10, use_numbers=True, use_symbols=False)
print("Generated Password:", pwd)
```

The Goal: Ask the user to input a long sentence or paragraph. Build a program that outputs statistics about their text.
The Catch: You must use separate functions for every analysis: count_words(text), count_vowels(text), and find_longest_word(text). Each function must take the text as an argument and return the result to be printed by the main program.
Concepts Tested: String manipulation, passing strings as arguments, returning specific values (integers and strings) from functions, and keeping logic strictly separated.

```python
def count_words(text):
    words = text.split()
    return len(words)
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
def find_longest_word(text):
    words = text.split()
    longest = ""
    for word in words:
        word = word.strip(".,!?;:")
        if len(word) > len(longest):
            longest = word
    return longest
def main():
    text = input("Enter a sentence or paragraph: ")

    word_count = count_words(text)
    vowel_count = count_vowels(text)
    longest_word = find_longest_word(text)

    print("\n\t TEXT STATISTICS ")
    print("Total words:", word_count)
    print("Total vowels:", vowel_count)
    print("Longest word:", longest_word)
    
main()
```
