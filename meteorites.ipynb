**The Scenario:** You act as junior seismologists analyzing recent global earthquake data.

- **Data Structures:** You separate out specific columns, like the magnitude, to see how it behaves as an isolated 1D Series compared to the 2D DataFrame context.
- **Data I/O:** You start by loading the data using pd.read_csv(). Later, you will export their filtered findings using .to_excel() or .to_json() for a web team to use.
- **Initial Data Inspection:** Before doing anything, you must run .info() and .describe() to understand the sheer volume of data, use .head() to see the layout, and use .value_counts() to see which regions have the most frequent seismic activity.

**Indexing and Selection:**
- Column/Label/Integer: Use dictionary-style selection to grab the 'Location' column. Use .iloc[] to extract the top 5 largest earthquakes based on row position.
- Filtering: Apply multiple boolean conditions using & and | to find earthquakes that are both greater than magnitude 6.0 and occurred in a specific region (e.g., "Japan" or “Texas”).

```python
import pandas as pd

df = pd.read_csv("meteorites.csv")

df.info()
df.describe()
df.head()

# df["classification"].value_counts()

fall_status = df["fall_status"]
print(fall_status.head())

df_sorted = df.sort_values(by="mass_grams", ascending=False)
top_5_heaviest = df_sorted.iloc[:5]
print(top_5_heaviest)

print("mass_grams > 50000:\n")
filtered = df[
    (df["mass_grams"] > 50000) &
    ((df["region"] == "Antarctica") | (df["region"] == "Sahara"))
]
print(filtered)

filtered.to_excel("heavy_meteorites.xlsx", index=False)
print("Saved Excel file")
filtered.to_json("heavy_meteorites.json", orient="records")
print("Saved JSON file")

```
