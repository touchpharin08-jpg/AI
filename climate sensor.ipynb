```python
import numpy as np
import matplotlib.pyplot as plt

climate_data = np.load("climate_data.npy")  

print("Shape:", climate_data.shape)   
print("Dimensions:", climate_data.ndim)
print("Data type:", climate_data.dtype)

baseline_grid = np.zeros(climate_data.shape[:2])

region = climate_data[20:50, 70:120, :]

day15 = climate_data[:, :, 14]  

clean_data = climate_data.astype(float)  
mask_invalid = clean_data == -99
clean_data[mask_invalid] = np.nan

fahrenheit = (clean_data * 9/5) + 32

heat_index = np.exp(clean_data / 30) + np.log(clean_data + 1)

max_temp = np.nanmax(clean_data)
min_temp = np.nanmin(clean_data)

print("Highest temperature:", max_temp)
print("Lowest temperature:", min_temp)

avg_temp_map = np.nanmean(clean_data, axis=2)

print("Average map shape:", avg_temp_map.shape)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("Day 15 Temperature Map")
plt.imshow(day15)
plt.colorbar(label="°C")

plt.subplot(1, 2, 2)
plt.title("Monthly Average Temperature")
plt.imshow(avg_temp_map)
plt.colorbar(label="°C")

plt.tight_layout()
plt.show()
```
