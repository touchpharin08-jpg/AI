The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
Concepts Tested: List methods (.append(), .pop()), indexing, iterating through lists with for loops.

```python
tasks = ["python", "java"]

while True:
    print("\n List")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Choose an option (1-4): ")
    if choice == "1":
        user_input = input("Enter option(add or done): ")
        if user_input == "done":
            break
        elif user_input == "add":
            task = input("Enter a new task: ")
            tasks.append(task)
            print("Task added")
            continue
    elif choice == "2":
        print("\nYour Tasks")
        if len(tasks) == 0:
            print("No tasks yet.")
        else:
            for i, task in enumerate(tasks):
                print(f"{i + 1}. {task}")

    elif choice == "3":
        if len(tasks) == 0:
            print("No tasks to remove.")
        else:
            for i, task in enumerate(tasks):
                print(f"{i + 1}. {task}")

            num = int(input("Enter task number to remove: "))
            if 1 <= num <= len(tasks):
                removed = tasks.pop(num - 1)
                print(f"Removed: {removed}")
            else:
                print("Invalid number.")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```

The Goal: Create a dictionary where the keys are questions (e.g., countries) and values are answers (e.g., capitals). Loop through the dictionary, ask the user the question, check their answer, and calculate a final score.
Concepts Tested: Dictionaries, key-value pairs, iteration, score tracking.

```python
quiz = {
    "What is the capital of France? ": "paris",
    "What is the capital of Japan? ": "tokyo",
    "What is the capital of Cambodia? ": "phnom penh"
}
score = 0
for question, answer in quiz.items():
    user_answer = input(question).lower()

    if user_answer == answer:
        print("Correct")
        score += 1
    else:
        print(f"Wrong! The correct answer is {answer}")
print("\n Quiz Finished")
print(f"Your score: {score}/{len(quiz)}")
```

The Goal: Ask the user to input the prices of items they are adding to their cart, one by one. Store these prices in a Python List. The loop stops when the user types 'done'. Then, calculate and print a receipt showing the total cost, the most expensive item in the cart, and the average item price.
Concepts Tested: while loops for input collection, list methods (.append()), built-in list functions (sum(), max(), len()), and data type casting (handling the mix of string commands and float prices).

```python
prices = []
while True:
    user_input = input("Enter item price (or type 'done'): ").lower()

    if user_input == "done":
        break
    price = float(user_input)
    prices.append(price)
if len(prices) > 0:
    total = sum(prices)
    highest = max(prices)
    average = total / len(prices)
    print("\n Receipt")
    print(f"Items: {len(prices)}")
    print(f"Total cost: ${total:.2f}")
    print(f"Most expensive item: ${highest:.2f}")
    print(f"Average price: ${average:.2f}")
else:
    print("No items were added")
```

The Goal: Create a dictionary pre-filled with cafe supplies and their quantities (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks the user if they want to "sell" or "restock" an item. Update the dictionary values based on their input and print the current inventory levels. Prevent the user from selling more items than they have in stock!
 Concepts Tested: Accessing and updating dictionary values, checking if a key exists using the in operator, combining if/elif logic with dictionary lookups, and managing state.

```python
inventory = {
    "coffee beans": 50,
    "cups": 100,
    "syrup": 15
}
while True:
    print("\n Cafe Inventory")
    for item, qty in inventory.items():
        print(f"{item}: {qty}")
    action = input("\nType 'sell', 'restock', or 'exit': ").lower()
    if action == "exit":
        print("Goodbye")
        break
    item = input("Enter item name: ").lower()
    if item not in inventory:
        print("Item not found in inventory.")
        continue
    quantity = int(input("Enter quantity: "))

    if action == "sell":
        if quantity <= inventory[item]:
            inventory[item] -= quantity
            print(f"Sold {quantity} {item}.")
        else:
            print("Not enough stock!")

    elif action == "restock":
        inventory[item] += quantity
        print(f"Restocked {quantity} {item}.")

    else:
        print("Invalid action.")
        
print("\n Final Inventory")
for item, qty in inventory.items():
    print(f"{item}: {qty}")
```
